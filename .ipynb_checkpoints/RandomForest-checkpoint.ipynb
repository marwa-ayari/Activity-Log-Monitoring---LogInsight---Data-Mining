```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Importer les données
df = pd.read_csv('Volume_Duration.csv', sep='\t')
print(df)
```

```python
#des informations sur la distribution des données dans chaque colonne numérique de ton DataFrame.
#Affichage du maximum, minimum et de la moyenne
df.describe()
```

```python
# visualiser les données contenues dans le DataFrame df a travers 
#un graphique plot de dimention scatter la durée (x) en fonction du volume (y)
#df.plot.scatter(x='Count', y='Duration')
```

```python
# Vérifier les valeurs aberrantes
q1 = df['Duration'].quantile(0.25)
q3 = df['Duration'].quantile(0.75)
iqr = q3 - q1
seuil_inf = q1 - 1.5 * iqr
seuil_sup = q3 + 1.5 * iqr
valeurs_aberrantes = df.loc[(df['Duration'] < seuil_inf) | (df['Duration'] > seuil_sup)]
print("Valeurs aberrantes :")
print(valeurs_aberrantes)
```

```python
# Remplacer les valeurs aberrantes par la médiane
mediane = df['Duration'].median()
seuil_inf = mediane - 3 * df['Duration'].std()
seuil_sup = mediane + 3 * df['Duration'].std()
df['Duration'] = np.where((df['Duration'] < seuil_inf) | (df['Duration'] > seuil_sup), mediane, df['Duration'])
```

```python
# Afficher l'histogramme des valeurs de la variable "Duration" après le remplacement
sns.histplot(data=df, x="Duration")
plt.title("Distribution des valeurs de la variable 'Duration' (APRES le remplacement)")
plt.show()
```

```python
#visualiser les donées avant la division du modéle d'entraînement
df.plot.scatter(x='Count', y='Duration')
```

```python
# Diviser les données en ensemble d'entraînement et de test
X = df.drop(['Duration'], axis=1)
y = df['Duration']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Visualiser la division des données graphiquement
# Créer une figure et un axe
fig, ax = plt.subplots()
# Tracer les points pour l'ensemble d'entraînement en bleu
ax.scatter(X_train['Count'], y_train, color='blue', label='Train')

# Tracer les points pour l'ensemble de test en rouge
ax.scatter(X_test['Count'], y_test, color='red', label='Test')

# Ajouter une légende et des titres d'axe
ax.legend()
ax.set_xlabel('Count')
ax.set_ylabel('Duration')
plt.show()
```

```python
# Initialiser le modèle Random Forest
rf_model = RandomForestRegressor(n_estimators=100)
```

```python
# Entraîner le modèle sur l'ensemble d'entraînement
rf_model.fit(X_train, y_train)
```

```python
# Prédire les valeurs de la durée pour l'ensemble de test
y_pred = rf_model.predict(X_test)
```

```python
# Calculer les métriques d'évaluation (RMSE et R2)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)
#affichage des résultats
print("RMSE: ", rmse)
print("R2 score: ", r2)
```

```python
# Prédire les valeurs de la durée pour l'ensemble des données
y_pred_all = rf_model.predict(X)
```

```python
# Calculer les métriques d'évaluation (RMSE et R2) pour l'ensemble des données
rmse_all = np.sqrt(mean_squared_error(y, y_pred_all))
r2_all = r2_score(y, y_pred_all)
#affichage des résultats
print("RMSE (ensemble des données): ", rmse_all)
print("R2 score (ensemble des données): ", r2_all)
```

```python
# Tracer la courbe de prédiction
plt.figure(figsize=(10,6))
plt.plot(y, label='Valeurs réelles')
plt.plot(y_pred_all, label='Valeurs prédites')
plt.legend()
plt.title('Comparaison des valeurs réelles et prédites')
plt.xlabel('Index des données')
plt.ylabel('Durée')
plt.show()
```
